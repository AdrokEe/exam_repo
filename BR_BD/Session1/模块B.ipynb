# 任务一

```python
# ans1_1
order = pd.read_excel("./模块B订单信息表.xlsx", sheet_name=0, dtype={"数量": int})
customer = pd.read_excel("./模块B客户信息表.xlsx", sheet_name=0)
```

```python
customer["客户等级"].unique()
```

```python
customer.info()
```

```python
ans1_2 = customer.query("`客户等级` == '高价值客户'")
ans1_2
```

```python
ans1_3 = customer[~customer.isin(ans1_2)].dropna()
```

```python
ans1_4 = ans1_3
ans1_4.loc[ans1_3["职业"] == "公司白领", "是否退休"] = ans1_4.loc[ans1_3["职业"] == "公司白领", "是否退休"].str.replace("否", "是")
ans1_4
```

```python
# ans 1_5
# order_time = order["订单日期"].astype("str").str.split("-")
# month = order_time.map(lambda x: x[1])
# month
month = order["订单日期"].dt.month
season = month.map(lambda x: "春" if x in [3, 4, 5] else "夏" if x in [6, 7, 8] else "秋" if x in [9, 10, 11] else "冬")
order.loc[:, "season"] = season
order
```

# 任务二

```python
ans2_1 = order.groupby("season").agg(订单数量=("订单编号", "count"))
ans2_1 = ans2_1.loc[ans2_1["订单数量"].idxmax(), :]
ans2_1
```

```python
order.loc[:, "是否打折"] = order["折扣"].apply(lambda x: "是" if x > 0 else "否")
ans2_2 = order.groupby("是否打折").agg(占比=("订单编号", "count"))
ans2_2 = ans2_2.transform(lambda x: x / x["占比"].sum()).round(2)
ans2_2
```

```python
ans2_3 = order.query("(`是否退回` == 0) and `实际配送天数` != 0")
ans2_3 = ans2_3.drop_duplicates(subset=["订单编号"])
ans2_3 = ans2_3["实际配送天数"].median()
ans2_3
```

```python
ans2_4 = order.loc[order["实际配送天数"] - order["计划配送天数"] < 0, :]
order.loc[order["实际配送天数"] - order["计划配送天数"] < 0, "实际配送天数"] = ans2_4["计划配送天数"] + 1
order
```

```python
order["产品名称"].unique()
```

```python
order.loc[:, "产品"] = order["产品名称"].str.split("_").str[1]
order.loc[:, "特征"] = order["产品名称"].str.split("_").str[-1]
ans2_5 = order.loc[order["特征"].str.startswith("红")].groupby("产品").agg(数量=("订单编号", "count"))
ans2_5.loc[ans2_5.idxmax()]
```

```python
order.loc[:, "订单地区年份"] = order["订单编号"].str.extract(r"(.*?-.*?)-.*?")
order.loc[order["订单编号"].str.startswith("CN"), "订单编号"] = order.loc[order["订单编号"].str.startswith("CN"), "订单编号"].str.replace("CN", "C")
ans3_1 = order.merge(customer, how="left", on="客户编号")
ans3_1
```

```python
ans3_2 = ans3_1.query("`是否退回` == 0").groupby("客户编号").agg(次数=("订单编号", "count"))
ans3_2 = ans3_2.loc[ans3_2["次数"] >= 20, :]
ans3_2
```
