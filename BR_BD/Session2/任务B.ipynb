```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("模块B数据.xlsx", index_col=0)
df.head()
```

```python
df_notnull = df.drop(df.loc[:, (df.isnull().sum(axis=0) / df.count() > 0.6)].columns, axis=1)
df_notnull.info()
df_notnull.to_csv("./results/1_2.txt", sep="\t")
```

```python
df_drop_duplicates = df_notnull.drop_duplicates()
df_drop_duplicates.info()
df_drop_duplicates.to_csv("./results/1_3.txt", sep="\t")
```

```python
df_clean = df_drop_duplicates.loc[df_drop_duplicates["利润额"] > 100, :]
df_clean.info()
df_clean.to_csv("./results/1_4.txt", sep="\t")
```

```python
df_clean.loc[:, "订单日期"] = df_clean.loc[:, "订单日期"].astype("str")
df_clean.loc[:, "day"] = pd.to_datetime(df_clean.loc[:, "订单日期"], format="%Y-%m-%d") + pd.Timedelta(hours=1, minutes=1, seconds=1)
df_clean.to_csv("./results/1_5_day.txt", sep="\t")
df_clean.loc[:, "week"] = df_clean.loc[:, "day"].dt.weekday + 1
df_clean.to_csv("./results/1_5_week.txt", sep="\t")
df_clean.loc[:, 'hour'] = df_clean.loc[:, "day"].dt.hour
df_clean.to_csv("./results/1_5_hour.txt", sep="\t")
# df_clean.loc[:, 'day'].dt.strftime('%I %p')
```

```python
# 创建价格区间
price_bins = [0, 1, 100, 500, 1000, 2000, 3000, 4000, 5000, 10000, float('inf')]
price_labels = ['0-1', '1-100', '101-500', '501-1000', '1001-2000', '2001-3000', '3001-4000', '4001-5000', '5001-10000', '10000以上']

# 使用 pd.cut 将价格字段划分到价格区间，并保存到 price_level 列
df_clean.loc[:, 'price_level'] = pd.cut(df_clean['利润额'], bins=price_bins, labels=price_labels, right=False)

result2_1 = df_clean.groupby("price_level").agg(次数=("price_level", "count"))
result2_1 = result2_1.loc[result2_1.idxmax()]
result2_1.to_csv("./results/2_1.txt", sep="\t", index=True)
```

```python
result2_2 = df_clean.loc[df_clean["price_level"] == result2_1.index[0], :]
result2_2 = result2_2.groupby("客户类型").agg(占比=("客户类型", "count"))
result2_2 = result2_2.transform(lambda x: x / x.sum()).round(4)
result2_2.to_csv("./results/2_2.txt", sep="\t")
```
